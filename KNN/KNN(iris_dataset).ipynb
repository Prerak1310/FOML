####<mark>Loading Dataset</mark>  

```python
import seaborn as sns
dataset = sns.load_dataset('iris')
dataset = dataset[['sepal_length','sepal_width','species']]
dataset = dataset.iloc[:100,:]
dataset
```

####<mark>Label Encoding Species</mark>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['species'] = le.fit_transform(dataset['species'])
dataset
```

####<mark>Dividing into 2 dataset as per flowers</mark>


```python

df0 = dataset[dataset.species==0]
df1 = dataset[dataset.species == 1]
```

####<mark> Plotting the Species </mark>


```python
from matplotlib import pyplot as plt
plt.scatter( df0.sepal_length,df0.sepal_width , color = 'red')
plt.scatter( df1.sepal_length,df1.sepal_width , color = 'green')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

####<mark>Splitting dataset </mark>

```python
from sklearn.model_selection import train_test_split
import numpy as np
X = dataset[['sepal_length','sepal_width']]
y = dataset.species
X_train, X_test , y_train, y_test = train_test_split( X , y , test_size = 0.3)
```

####<mark>fitting the model </mark>


```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 10)
model.fit(X_train,y_train)
```

####<mark>Score of model </mark>


```python
pred_values = model.predict(X_test)
print(model.score(X_test, y_test))
```

####<mark>Confusion Matrix</mark>


```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred_values )
```
