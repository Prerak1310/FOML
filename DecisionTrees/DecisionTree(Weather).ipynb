####<mark> dataset </marK>

```python
import pandas as pd
weather_dataset = pd.read_csv('/content/dataset.csv')
weather_dataset
```

####<mark>LabelEncoding</mark>

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
weather_dataset=weather_dataset.apply(le.fit_transform)

weather_dataset
```

```python
feature_columns = ['outlook','humidity',"windy"]
```

```python
import numpy as np
X = weather_dataset[feature_columns]
y = weather_dataset['play']

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size= 0.2)
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='')
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn.tree import plot_tree
plot_tree(classifier)
```
