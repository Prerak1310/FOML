##<mark>Importing dataset</mark>

```python
from sklearn.datasets import load_digits
digits_dataset = load_digits()
digits_dataset
```

####<mark>various attributes of digits dataset</mark>

```python
dir(digits_dataset)
```

####<mark>Images in our dataset</mark>

```python
from matplotlib import pyplot as plt
plt.matshow(digits_dataset.images[0])
plt.matshow(digits_dataset.images[5])
```

####<mark>Adding target column to dataframe</mark>

```python
import pandas as pd
df = pd.DataFrame(digits_dataset.data , columns=digits_dataset.feature_names)
df['target'] = digits_dataset.target
df
```

####<mark></mark>

##<mark>Train Test Split </mark>

```python
from sklearn.model_selection import train_test_split
X= df.drop(['target'],axis="columns")
y= df.target
X_train,X_test,y_train,y_test = train_test_split( X,y ,test_size = 0.2)
```

##<mark> Model Fitting and Score Prediction</mark>

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 30) #Estimators are the number of trees
model.fit(X_train,y_train)
```

####<mark>Score and Predictions</mark>

```python
predicted_values = model.predict(X_test)
model.score(X_test,y_test)
```

####<mark>Confusion Matrix</mark>


```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predicted_values)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize = (10,7))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
```
